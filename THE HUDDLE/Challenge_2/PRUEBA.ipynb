```python
# Demostración de Herencia en Python

La **herencia** permite crear nuevas clases basadas en clases existentes, heredando sus atributos y métodos.

## Conceptos clave:
- **Clase padre (superclase)**: La clase base de la cual se hereda
- **Clase hija (subclase)**: La clase que hereda de otra
- **super()**: Función para acceder a métodos de la clase padre
```

```python
# Ejemplo 1: Clase Padre (Animal)
class Animal:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad
        print(f"Animal creado: {nombre}")
    
    def hacer_sonido(self):
        return "Algún sonido genérico"
    
    def dormir(self):
        return f"{self.nombre} está durmiendo"
    
    def info(self):
        return f"Nombre: {self.nombre}, Edad: {self.edad} años"

# Creamos un animal genérico
animal = Animal("Rex", 3)
print(animal.info())
print(animal.hacer_sonido())
```
